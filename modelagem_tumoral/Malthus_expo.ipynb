Modelo de crescimento exponencial de Malthus

Hipóteses:

* Considera-se um tumor formado por um único tipo de célula;
* Sem variação espacial, não há migração;
* Sem fontes de nutrientes;

A QoI a ser estudada é a densidade de células de câncer no tempo t: **N(t)**, em que t é a variável independente e N a variável dependente.

A equação base é dada por: 
$N(t) = N_0 * e^(r*t)$

```python
#bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def malthus_growth_model(N0:float, r:float, t:list[int])->list[float]:
    '''
    Malthus growth model
    
    Arguments:
        N0: initial population
        r: growth rate
        t: time
    
    Returns:
        N: population at time t

    '''
    return N0 * np.exp(r * t)
```

```python
# Set parameters
N0 = 2  # Initial population
r = 0.1  # Growth rate
t = np.arange(0, 50, 1)  # Time values from 0 to 50
```

```python
# Calculate population values
population = malthus_growth_model(N0, r, t)
```

```python
# Plot the results
plt.plot(t, population, label='Malthusian Growth Model')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Malthusian Growth Model')
plt.legend()
plt.show()
```

Pode-se relacionar o Modelo de Malthus ao crescimento do tumor apenas do
início do seu desenvolvimento, até um determinado período, em que todas as células ainda
encontrem recursos para suprir suas necessidades de sobrevivência e, provavelmente, este
período terminará antes que ocorra a angiogênese.

```python
def malthus_growth_trap_model(N0:float, r_population:float, r_resources:float, R0:float, t:list[int])->list[float]:
    '''
    Malthus growth trap model
    
    Arguments:
        N0: initial population
        r_population: population growth rate
        r_resources: resources growth rate
        R0: initial resources
        r: growth rate
        t: time
    
    Returns:
        population: population at time t
        resources: resources at time t

    '''
    population = N0 * np.exp(r_population * t)
    resources = r_resources * t + R0

    return population, resources
```

```python
# Set parameters
N0 = 2
R0 = 10 
r_resources = 2
r_population = 0.1
t = np.arange(0, 50, 1)  # Time values from 0 to 50
```

```python
#Calculate population and resources values
population, resources = malthus_growth_trap_model(N0, r_population, r_resources, R0, t)
```

```python
#pllot the results
plt.plot(t, population, label='Population')
plt.plot(t, resources, label='Resources')
plt.xlabel('Time')
plt.ylabel('Population/Resources')
plt.title('Malthusian Growth Trap Model')

#Intersection points
idx = np.argwhere(np.diff(np.sign(population - resources))).flatten()
plt.plot(t[idx], resources[idx], 'ro', label='Point of crisis')
# print(f'Intersection point: {intersection}')

plt.legend()
plt.show()

```
